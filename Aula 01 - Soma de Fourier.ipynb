```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-np.pi,np.pi,100)
f = np.zeros(len(t))
f[t>0]=1
```

```python
plt.plot(t,f)
```

```python
f0 = np.ones(len(t))*1/2
f1 = f0 + 2/(np.pi)*np.sin(t)
f2 = f1 + 2/(3*np.pi)*np.sin(3*t)
f3 = f2 + 2/(5*np.pi)*np.sin(5*t)
```

```python
plt.plot(t,f,'o')
plt.plot(t,f0,'--r')
plt.plot(t,f1,'--g')
plt.plot(t,f3,'-or')
```
